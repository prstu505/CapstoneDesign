```python
# >>> import library <<<
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import WebDriverException
import time
import os
import json


# >>> set chrome_driver options <<<
chrome_options = Options()
chrome_options.add_argument('--mute-audio') # disable audio
chrome_options.add_argument('--disable-features=Translate') # disable auto-translation of chrome
# chrome_options.add_argument("headless") # do not show window
chrome_options.add_experimental_option("detach", True) # keep browser open after script is completed


# >>> declare constants <<<
PATHS = ['./data/modern essay_sub', './data/modern novel_sub', './data/modern poem_sub']
KINDS = ['essay', 'novel', 'poem']


# >>> declare functions <<<


# create chrome instance with options
driver = webdriver.Chrome(options=chrome_options)

# open translate page
driver.get(url='http://www.davincimap.co.kr/davServ/TextProc/davTransOldKor.jsp')

# translate and save each file
for i in range(len(PATHS)) :
    fileNameList = os.listdir(PATHS[i]) # get all file name of target directory
    # get file's content and save to rows
    for j in range(len(fileNameList)) :
        if fileNameList[j] in ['SOUR002315&&&셔유견문록 (西遊見聞錄).txt', 'SOUR008296&&&차창(차창).txt'] :
            continue
        
        file = open(PATHS[i] + '\\' + fileNameList[j], 'r', encoding='utf-8')
        row = {}
        count = 0
        while True : 
            line = file.readline()
            line = line.replace('\n', '')

            # check EOF
            if not line : break

            if count == 0 : 
                row['sourid'] = line.split(':')[1]
            elif count == 1 : 
                row['title'] = line.split(':')[1]
            elif count == 2 : 
                row['author'] = line.split(':')[1]
            elif count == 3 : 
                pass
            elif count == 4: 
                row['content'] = line
            count += 1
        file.close()

        # split file's content
        splitContent = []
        ceil = 500
        while ceil < len(row['content']) : 
            splitContent.append(row['content'][ceil-500:ceil])
            ceil += 500
        splitContent.append(row['content'][ceil-500:len(row['content'])])

        # log
        print('<' + str(j) + '>\n' + row['title'] + row['author'])

        # change old korean to modern korean using davinchimap tool
        for k in range(len(splitContent)) : 
            try : 
                WebDriverWait(driver, 5).until(EC.presence_of_element_located((By.XPATH, '/html/body/table[3]/tbody/tr/td[4]/form/table[2]/tbody/tr[1]/td[2]/textarea'))).clear()
                WebDriverWait(driver, 5).until(EC.presence_of_element_located((By.XPATH, '/html/body/table[3]/tbody/tr/td[4]/form/table[2]/tbody/tr[1]/td[2]/textarea'))).send_keys(splitContent[k])
                time.sleep(0.5) # for reducing server load
                WebDriverWait(driver, 5).until(EC.presence_of_element_located((By.XPATH, '/html/body/table[3]/tbody/tr/td[4]/form/table[2]/tbody/tr[3]/td[2]/button'))).click()
                splitContent[k] = WebDriverWait(driver, 5).until(EC.presence_of_element_located((By.XPATH, '/html/body/table[3]/tbody/tr/td[4]/table[2]/tbody/tr[2]/td[2]/textarea'))).text
            except WebDriverException :
                j -= 1
                driver.get(url='http://www.davincimap.co.kr/davServ/TextProc/davTransOldKor.jsp')
                break

        # combine all results into one string
        res = ''
        for e in splitContent : 
            res = res + e
        row['content'] = res
        
        time.sleep(1) # for reducing server load

        # save result
        with open('./data/modern_' + KINDS[i] + '_3/' + row['sourid'] + '&&&' + row['title'] + '.json', 'w', encoding='utf-8') as file :
            json.dump(row, file, indent='\t', ensure_ascii=False)

driver.close()
```
