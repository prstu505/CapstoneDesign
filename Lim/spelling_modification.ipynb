```python
import re
import requests

def get_passport_key():
    url = "https://search.naver.com/search.naver?where=nexearch&sm=top_hty&fbm=0&ie=utf8&query=네이버+맞춤법+검사기"
    res = requests.get(url)

    html_text = res.text

    match = re.search(r'passportKey=([^&"}]+)', html_text)
    if match:
        passport_key = match.group(1)
        return passport_key
    else:
        return False

def fix_spell_checker_py_code(file_path, passportKey):
    pattern = r"'passportKey': '.*'"

    with open(file_path, 'r', encoding='utf-8') as input_file:
        content = input_file.read()
        modified_content = re.sub(pattern, f"'passportKey': '{passportKey}'", content)

    with open(file_path, 'w', encoding='utf-8') as output_file:
        output_file.write(modified_content)
    
    return 

spell_checker_file_path = './hanspell/spell_checker.py'
passport_key = get_passport_key()
if passport_key:
    fix_spell_checker_py_code(spell_checker_file_path, passport_key)
else:
    print("passportKey를 찾을 수 없습니다.")
```

```python
from hanspell import spell_checker
import os
import json

SRC_PATH = ['./data/3_modern_essay', './data/3_modern_novel', './data/3_modern_poem']
DST_PATH = ['./data/4_modern_essay', './data/4_modern_novel', './data/4_modern_poem']

for i in range(len(SRC_PATH)) :
    src_path = SRC_PATH[i]
    dst_path = DST_PATH[i]
    file_list = os.listdir(src_path)
    for j in range(0, len(file_list), 1) :
        file_name = file_list[j]

        try :
            print('##### start #####\t' + str(j) + '\t' + file_name)

            book = None
            with open(src_path+'/'+file_name, 'r', encoding='utf-8') as file :
                book = json.load(file)
                
                splitContent = []
                ceil = inter = 300
                while ceil < len(book['content']) :
                    splitContent.append(book['content'][ceil-inter:ceil])
                    ceil += inter
                splitContent.append(book['content'][ceil-inter:len(book['content'])])
                
                result = []
                for content in splitContent :
                    result.append(spell_checker.check(content).checked)

                book['content'] = ''.join(result)

            with open(dst_path+'/'+file_name, 'w', encoding='utf-8') as file :
                json.dump(book, file, indent='\t', ensure_ascii=False)
            
            print('#####  end  #####\t' + str(j) + '\t' + file_name)
        except :
            print('>>> error <<<\t' + file_name)
```
