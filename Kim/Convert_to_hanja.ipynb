```python
import json
import hanja

Similar_data = {"data": []}
dict = {"data": [{"para": [{"label": [{}], "context": ""}]}]}

with open("./Similar_large_noduplicate_kobert_data.json", 'rb') as f :
    data = json.load(f)

for info in data["data"]:
    for form in info["para"]:
        para = {"para": []}
        context = form["context"]
        labels = {"label":[]}
        label ={}
        for content in form["label"]:
            sequence = hanja.translate(content["sequence"], 'substitution')
            token_str = hanja.translate(content["token_str"], 'substitution')
            label["sequence"] = sequence
            label["token_str"] = token_str
            labels["label"].append(label)
        labels["context"]= form["context"]
        para["para"].append(labels)
    Similar_data["data"].append(para)

with open('./Similar_large_robert_data_modify.json', 'w', encoding='utf-8') as file :
    json.dump(Similar_data, file, indent='\t', ensure_ascii=False)
```
