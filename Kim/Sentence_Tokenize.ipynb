```python
import glob
import os
import json

file_paths_1 = glob.glob('data_final/modern_essay_3/*.json')
file_paths_2 = glob.glob('data_final/modern_novel_3/*.json')
file_paths_3 = glob.glob('data_final/modern_poem_3/*.json')

file_contents = []

def load_json (file_contents,file_paths):
    for file_path in file_paths:
        with open(file_path, 'r', encoding='utf-8') as file:
            content = json.load(file)
            file_contents.append(content)
           

load_json(file_contents,file_paths_1)
load_json(file_contents,file_paths_2)
load_json(file_contents,file_paths_3)
```

```python
import pandas as pd
import numpy as np
df = pd.DataFrame(file_contents)
df.head(5)

john_row = df.loc[df['title'] == ' 진달래꽃 (시집)']

print(john_row)
```

```python
import konlpy 
from konlpy.tag import Okt, Hannanum, Kkma 

from kss import split_sentences

import nltk
from nltk.tokenize import sent_tokenize
from kiwipiepy import Kiwi

poem_text = df["content"][4526]
kiwi = Kiwi()
okt = Okt()
kkma = Kkma()
hannanum= Hannanum()
novel_text = df["content"][0]

kss_sentences = split_sentences(poem_text)
kiwi_sentences = kiwi.split_into_sents(poem_text)
kss_sentences1 = split_sentences(novel_text)
kiwi_sentences1 = kiwi.split_into_sents(novel_text)
def sentence_tokenizer (sentences):
    print(sentences)
    # for i in sentences:
    #     print(i)
    print("문장 갯수: ",len(sentences),"\n")

print(poem_text)

print("온점(구분점)이 없는 poem 종류의 글")
print("----nltk----\n")

print("----kss----\n")
sentence_tokenizer(kss_sentences)

print("----kiwi----\n")
sentence_tokenizer(kiwi_sentences)

print("온점(구분점)이 있는 essay, novel 이 있는 글\n\n\n")
print("----nltk----")

print("----kss----\n")
sentence_tokenizer(kss_sentences1)

print("----kiwi----\n")
sentence_tokenizer(kiwi_sentences1)
```

```python
import tqdm
content_list = df.to_numpy().tolist()
sentence_token = pd.DataFrame(columns=['content'])

# tqdm을 enumerate 함수와 함께 사용
for i, text in enumerate(content_list):
    sen_token = kiwi.split_into_sents(text[0])
    if i == 10:
        break
    for token in sen_token:
        new_row = pd.DataFrame({'content': [token[0]]})
        sentence_token = sentence_token.append(new_row, ignore_index=True)
        
```
